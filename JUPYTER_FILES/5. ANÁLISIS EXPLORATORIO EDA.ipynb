```python
#% Valores nulos
total = data.isnull().sum().sort_values(ascending=False)
percent_1 = data.isnull().sum()/data.isnull().count()*100
percent_2 = (round(percent_1, 1)).sort_values(ascending=False)
missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
missing_data 

#No hay valores nulos, ya que se imputaron
```

```python
#Generamos el report de pandas con más información
desc_min = ProfileReport(data, minimal=True)
desc_min.to_file(output_file="Report_pandasprofile.html")
```

```python
#Genereamos el reporr de sweetviz
np.warnings = warnings
np.bool = bool
my_report = sweetviz.analyze([data, "Datos alumnado 1FPB"],target_feat='PROMOCIONA')
my_report.show_html('ReportSweetviz.html')
```

```python
#Tabla cruzada entre idioma del alumno y si promociona
data_crosstab = pd.crosstab(data.PROMOCIONA, data.A_IDIOMA_NVL, margins = False)
data_crosstab
```

```python
#Análisis sobre la variable objetivo

#EDA (Análisis Exploratorio de Datos)
df=data
#1)Tabla de frecuencias para el df:
#Total registros df -1750
reg_tot = len(df)
#Datos únicos (clases)
lis_class = df['PROMOCIONA'].unique()
#Creamos el dataframe que contendrá las clases
data = pd.DataFrame(lis_class, index=[1, 2], columns=["PROMOCIONA"] )
data=data.sort_index(ascending=True)

#Obtenemos las frecuencias absolutas de cada clase
data_ni = pd.crosstab(index=df["PROMOCIONA"], columns = "ni").sort_values('ni',ascending=True)
#Creamos una lista con las f absolutas
lis_ni = data_ni.values
#Agregamos la colmna de frecuencias absolutas
data['ni']=lis_ni

#Tabla de frecuencia relativa
data_fi = data_ni["ni"] / reg_tot
data_fi = data_fi.values
#Agregamos nueva columna de frecuentas relativas
data["fi"] = data_fi

#Obtenemos las frecuencias absolutas acumuladas
#Sacamos una lista de los valores donde obtendremos la FI
Ni = data["ni"].values

#Recorremos la lista para ir creando una nueva lista con las sumas
#Obtener la frecuencia absoluta acumulada
a = []
b = 0
for c in Ni:
    b = c + b
    a.append(b)
#Agregamos la nueva columna Fi al Dataframe dat
data["Ni"] = a

#Recorremos lista para obtener la frecuencia relativa acumulada
Hi = data["fi"].values
#Obtenemos Hi
a = []
b = 0
for c in Hi:
    b = c + b
    a.append(b)

#Agregamos la nueva columna Hi al Dataframe
data["Fi"] = a

comment='''
Al ser categórica, la Ni y Fi no tienen sentido, por tanto
sólo nos aportan información la ni (¿Cuantos hay?) y fi(¿Qué proporción hay?)
'''
data
```

```python
#Grafico en una ventana aparte de la proporción para cada clase

import matplotlib.pyplot as plt #Representación de gráficos

plt.switch_backend('TkAgg')
#Configuramos etiquetas y tamaños
labels = data["PROMOCIONA"]
sizes = data["ni"]
#Dibujamos el gráfico
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels = labels,
autopct = '%1.1f%%', startangle = 90)
ax1.axis('equal')
#Titulo eje x
plt.xlabel("PROMOCIONA EL ALUMNO 0(N) - 1(S)")
#Tamaño
plt.rcParams["figure.figsize"] = (18,6)
#Mostramos
plt.show()

comment='''Vemos la descompensación de las clases en el dataset de forma visual'''
```
